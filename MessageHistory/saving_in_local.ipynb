```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_ollama import ChatOllama
# Define LLM instances
llama_llm = ChatOllama(base_url="http://localhost:11434",
                       model="qwen2.5:latest", 
                       temperature=0.7,
                       max_tokens=250)
```

```python
from langchain_core.prompts import ChatPromptTemplate
# Define a chat prompt template with placeholders for history and prompt
message_prompt = ChatPromptTemplate.from_messages(
    [("placeholder", "{history}"),
    ("human", "{prompt}")]
)
```

```python
from langchain_core.output_parsers import StrOutputParser
# Define the chain using the prompt, LLM, and output parser
chain = message_prompt | llama_llm | StrOutputParser()
```

```python
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
# In-memory session store for demonstration purposes
session_store = {}
session_id = "user_session_123"
def get_session_history(session_id: str) -> BaseChatMessageHistory:
    """Retrieve or create a chat message history for the given session ID."""
    # For simplicity, we use an in-memory history. In production, this could be a database or file storage.
    if session_id not in session_store:
        session_store[session_id] = ChatMessageHistory()
    return session_store[session_id]
BaseChatMessageHistory.clear(session_store.get(session_id))
# Wrap the chain with message history management
history = RunnableWithMessageHistory(chain,
                                     get_session_history,
                                     input_messages_key="prompt",
                                     history_messages_key="history")
```

```python
# Clear previous history for the session
get_session_history(session_id).clear()
# Example invocations with session ID to maintain history
history.invoke({"prompt": "Explain the concept of LangChain in brief."},
               config={"session_id": session_id})
history.invoke({"prompt": "How does it help in building applications?"},
               config={"session_id": session_id})
```
