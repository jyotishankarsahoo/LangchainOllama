```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_ollama import ChatOllama
local_llm = ChatOllama(base_url="http://localhost:11434",
                       model="qwen2.5:latest",
                       temperature=0.5,
                       max_tokens=250)
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("placeholder","{history}"),
        ("human"),"{prompt}"
    ]
)
```

```python
from langchain_core.output_parsers import StrOutputParser
chain = prompt | local_llm | StrOutputParser()
```

```python
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import SQLChatMessageHistory

def get_session_data(session_id) -> BaseChatMessageHistory:
    return SQLChatMessageHistory(session_id=session_id, connection_string="sqlite:///chatHistory.db")
history = RunnableWithMessageHistory(chain,
                                     get_session_history=get_session_data,
                                     input_messages_key="prompt",
                                     history_messages_key="history")
session_id = "1021325"

response1 = history.invoke({"prompt": "What is capital of france?"},
                           config={"session_id": session_id})
response2 = history.invoke({"prompt": "What famous in it?"},
                           config={"session_id": session_id})
print(response1)
print("\n\n")
print(response2)
```
